```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from functools import reduce
from bs4 import BeautifulSoup
from pathlib import Path
from talib import SMA
import yfinance as yf
import seaborn as sns
import pandas as pd
import requests
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set_style("whitegrid")
idx = pd.IndexSlice
```

```python


def plot_stock_mv (data, ticker_):
    fig = make_subplots(1,1, subplot_titles = (f"{ticker_} Stock Movement",""))
    fig.append_trace(go.Candlestick(x = data.index, open=data.loc[:, idx["Open", ticker_]],
                                    high = data.loc[:, idx["High", ticker_]],
                                    low = data.loc[:, idx["Low", ticker_]],
                                    close= data.loc[:, idx["Close", ticker_]],
                                    increasing_line_color='#22FD55',decreasing_line_color='red',showlegend=False),row=1, col=1)

    # SM50
    fig.append_trace(go.Scatter
                     (x=data.index,
                      y =  data.loc[:, idx["SM200", ticker_]], 
                      line=dict(color='red', width=2),
                      name = "SM200"), row=1, col=1)
    
    # SM50
    fig.append_trace(go.Scatter
                     (x=data.index,
                      y =  data.loc[:, idx["SM50", ticker_]], 
                      line=dict(color='#FFE600', width=2),
                      name = "SM50"), row=1, col=1)
    #SM30
    fig.append_trace(go.Scatter
                     (x=data.index,
                      y =  data.loc[:, idx["SM13", ticker_]], 
                      line=dict(color='#27B51D', width=2),
                      name = "SM13"), row=1, col=1)
    fig.append_trace(go.Scatter
                     (x=data.index,
                      y =  data.loc[:, idx["SM26", ticker_]], 
                      line=dict(color='#B5221D', width=2),
                      name = "SM26"), row=1, col=1)
    fig.append_trace(go.Scatter
                     (x=data.index,
                      y =  data.loc[:, idx["SM30", ticker_]], 
                      line=dict(color='#BF18D1', width=2),
                      name = "SM30"), row=1, col=1)
    fig.show()
```

<p> As I mentioned in the Index , this is for all of those who want to run the code and choose the starting-end day, or if you want to analyze the same data but in terms of week, year etc., however, you have to deal with some pandas transformation </p>

```python
# ---> Settings
url = 'https://www.slickcharts.com/sp500'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

# ---> Connecting
request = requests.get(url ,headers=headers)
bs = BeautifulSoup(request.content,"html5")
sp500 = bs.find('div', {'class':'table-responsive'})
Csp500 =[ticker.text for ticker in sp500.select('td~ td+ td a') if ticker.text not in ['BRK.B','BF.B']]
print(f"This are some of the tickers in the Sp500 \n{Csp500[1:10]}")

# ---> Downloading 
start, end = '2021-01-01' , '2022-04-11'
stocks_ = (yf.download(Csp500, start = start, end= end ,group_by="ticker",period ='1d' ))
```

<p> The Original Data shows tickers as the first order columns but I want them in the second order, therefore I have to sawplevel , unstack and many more </p>

```python
stocks_.head(7)
```

---------------------------------------------------------
<p> In this way I can easily Iterate between values : Adj Close, Close etc </p>

```python
pstocks = stocks_.copy()
pstocks = pstocks.swaplevel(axis=1).T.unstack().swaplevel(axis=1).T.swaplevel().unstack()

#Showing all Levels
pstocks.loc[:,idx[:,["AAPL", 'GOOGL']]].tail(7)
```

```python
!git clone https://github.com/raudez77/Spyzer.git
import Spyzer
from Spyzer.movingaverage import MovingAverage
```

```python
# ---> Initiating Class
MA = MovingAverage(pstocks) 
```

<li> The sub-function Multi_SMA calcualte different Moving Average and return a dataframe containing all calculations</li>
Calculate multi daily average: pass a list

```python
# ---> Calculating Moving Average
Mov_Ave = [200,50,30,26,13]
col = 'Adj Close'
MDMA = MA.Multi_SMA(Mov_Ave,col)
MDMA.tail(7)
```

Now lets joins Both Data set and Plot some graph

```python
all_data = pd.merge(pstocks, MDMA,left_index=True,right_index=True)
start, end = "2022-04-04","2022-04-08"
all_data.loc[start:end,idx[:,"AAPL"]]
```

```python
plot_stock_mv(all_data, 'AAPL')
```

```python
plot_stock_mv(all_data, 'AMZN')
```

 <p> The other function that I wrote is able to filter all those stock above 200, 50 which are clearly in trend , <b> have the wind on your back </b>, this function will need the close price/adj price <b> up to you</b> the starting day: filter since any time , and td, which is how many days has the stock in positive trend</p>.
 
 <li> I will start from January 01 , before the war in Ukraine started, I want find some stocks with positive trending or assume they has not been affected by the war</li>
 <li> Only Stocks that has been more than 50 days in positive trend (more than 2 month)</li>

```python
# ---> Filtering Stocks 
col = 'SM50'
start = '2022-01-02'
trend_day = 50

MA = MovingAverage(all_data) 
above_50 = MA.above_50(col_name = col, start= start,td = trend_day)
above_50.head(7)
```
